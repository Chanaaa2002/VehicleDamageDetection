```python
from pathlib import Path
import zipfile

zip_path = Path("/content/drive/MyDrive/Vehicle_Damage_Detection/data/processed/archive2.zip")
base = zip_path.parent
out_dir = base / "archive2"   # folder after extract

if not out_dir.exists():
    with zipfile.ZipFile(zip_path, "r") as z:
        z.extractall(base)

print("Done:", out_dir.exists(), "->", out_dir)
```

```python
from pathlib import Path

DATA_DIR = Path("/content/drive/MyDrive/Vehicle_Damage_Detection/data/processed/archive2")

print("DATA_DIR exists:", DATA_DIR.exists())
print("train exists:", (DATA_DIR/"train").exists())
print("val exists:", (DATA_DIR/"val").exists())

def count_imgs(p):
    exts = {".jpg", ".jpeg", ".png"}
    return sum(1 for f in p.rglob("*") if f.suffix.lower() in exts)

if (DATA_DIR/"train").exists():
    print("Train images:", count_imgs(DATA_DIR/"train"))
if (DATA_DIR/"val").exists():
    print("Val images:", count_imgs(DATA_DIR/"val"))
```

```python
!pip -q install ultralytics

from ultralytics import YOLO

model = YOLO("yolo11n-cls.pt")  # small + fast
results = model.train(
    data=str(DATA_DIR),   # IMPORTANT: points to folder with train/val
    imgsz=224,
    epochs=10,
    batch=32,
    name="archive2_yolo11n_cls"
)
```
