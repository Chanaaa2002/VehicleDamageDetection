```python
from pathlib import Path

base = Path(r"C:\Users\User\Desktop\Vehicle_Damage_Detection\data\processed")

for ds in ["cardd_coco", "cardd_sod"]:
    p = base / ds
    print(f"\n===== {ds} =====")
    if p.exists():
        for item in p.iterdir():
            if item.is_dir():
                print(f"\n[{item.name}]")
                for sub in list(item.iterdir())[:10]:
                    print("  -", sub.name)
```
